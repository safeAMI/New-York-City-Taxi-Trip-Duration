```python
def scatterplot(x, y, title, ax, marker, c, labels):
    ax.set_ylim(30, 60)                        #latitude
    ax.set_xlim(-130, -50)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title(title)
    ax.scatter(x, y, marker = marker, c = c, alpha = 0.3, label = labels)
    ax.legend(loc = 'best')
```

```python
fig = plt.figure(figsize = (12, 9))
ax_1 = fig.add_subplot(221)
train_df['pickup_latitude'].hist(bins = 5, ax = ax_1)
ax_1.set_title('Distribution of pickup latitude')

ax_2 = fig.add_subplot(222)
train_df['pickup_longitude'].hist(bins = 5, ax = ax_2)
ax_2.set_title('Distribution of pickup longitude')

ax_3 = fig.add_subplot(223)
train_df['dropoff_latitude'].hist(bins = 5, ax = ax_3)
ax_3.set_title('Distribution of dropoff latitude')

ax_4 = fig.add_subplot(224)
train_df['dropoff_longitude'].hist(bins = 5, ax = ax_4)
ax_4.set_title('Distribution of dropoff longitude')

#plt.subplots_adjust(wspace = 2, hspace = 2)
```

```python
%matplotlib inline
fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (9, 6))
scatterplot(train_df['pickup_longitude'],train_df['pickup_latitude'],"locations", ax, '<', 'r','train_pick')
scatterplot(train_df['dropoff_longitude'],train_df['dropoff_latitude'],"locations", ax, '>', 'r','train_drop')
scatterplot(test_df['pickup_longitude'],test_df['pickup_latitude'],"locations", ax, '<', 'g', 'test_pick')
scatterplot(test_df['dropoff_longitude'],test_df['dropoff_latitude'],"locations", ax, '>', 'g', 'test_drop')
```

```python
train_df['pickup_datetime'] = pd.to_datetime(train_df.pickup_datetime)
test_df['pickup_datetime'] = pd.to_datetime(test_df.pickup_datetime)
train_df.loc[:, 'pickup_date'] = train_df['pickup_datetime'].dt.date
test_df.loc[:, 'pickup_date'] = test_df['pickup_datetime'].dt.date
train_df['dropoff_datetime'] = pd.to_datetime(train_df.dropoff_datetime)
```

```python
fig = plt.figure(figsize = (12, 9))
ax_1 = fig.add_subplot(221)
y_label = train_df['trip_duration']
trip_duration_weekday = pd.concat([train_df['pickup_datetime'], y_label], axis = 1)
trip_duration_weekday['day'] = trip_duration_weekday["pickup_datetime"].dt.day
trip_duration_mean = trip_duration_weekday.groupby('day')['trip_duration'].mean()
trip_duration_mean.plot(ax = ax_1, marker = 'o', ls = '--')
ax_1.set_title("Mean duration of trips for each day of the month")

ax_2 = fig.add_subplot(222)
y_label = train_df['trip_duration']
trip_duration_weekday = pd.concat([train_df['pickup_datetime'], y_label], axis = 1)
trip_duration_weekday['weekday'] = trip_duration_weekday["pickup_datetime"].dt.weekday
trip_duration_mean = trip_duration_weekday.groupby('weekday')['trip_duration'].mean()
trip_duration_mean.index = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
trip_duration_mean.plot(ax = ax_2, marker = 'o', ls = '--')
ax_2.set_title("Mean duration of trips for each day of the week")

ax_3 = fig.add_subplot(223)
trip_duration_weekday = pd.concat([train_df['pickup_datetime'], y_label], axis = 1)
trip_duration_weekday['Hour'] = trip_duration_weekday["pickup_datetime"].dt.hour
trip_duration_mean = trip_duration_weekday.groupby('Hour')['trip_duration'].mean()
trip_duration_mean.plot(ax = ax_3, marker = 'o', ls = '--')
ax_3.set_title("Mean duration of trips for each hour of the day")

ax_4 = fig.add_subplot(224)
trip_duration_weekday = pd.concat([train_df['pickup_datetime'], y_label], axis = 1)
trip_duration_weekday['Month'] = trip_duration_weekday["pickup_datetime"].dt.month
trip_duration_mean = trip_duration_weekday.groupby('Month')['trip_duration'].mean()
trip_duration_mean.plot(ax = ax_4, marker = 'o', ls = '--')
ax_4.set_title("Mean duration of trips for each Month of the year")
```

```python
all_df = pd.concat((train_df,test_df), axis = 0)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=42).fit(all_df[['pickup_longitude','pickup_latitude']])
pickup_clusters = kmeans.predict(all_df[['pickup_longitude','pickup_latitude']])
kmeans = KMeans(n_clusters=5, random_state=42).fit(all_df[['dropoff_longitude','dropoff_latitude']])
dropoff_clusters = kmeans.predict(all_df[['dropoff_longitude','dropoff_latitude']])
```

```python
all_df['pickup_clusters'] = pickup_clusters
all_df['dropoff_clusters'] = dropoff_clusters
```

```python
%matplotlib inline
fig, ax = plt.subplots(ncols = 2, nrows =1 , figsize = (11, 4))
ax[0].scatter(all_df['pickup_longitude'],all_df['pickup_latitude'], c = all_df['pickup_clusters'])
ax[0].set_title = "pickup_clusters"
ax[1].scatter(all_df['dropoff_longitude'],all_df['dropoff_latitude'], c = all_df['dropoff_clusters'])
ax[1].set_title = "dropoff_clusters"
```
